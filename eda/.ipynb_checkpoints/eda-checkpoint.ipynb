```python
df['sex'] = df['sex'].astype('category')
df['cp'] = df['cp'].astype('category')
df['fbs'] = df['fbs'].astype('category')
df['restecg'] = df['restecg'].astype('category')
df['exang'] = df['exang'].astype('category')
df['slope'] = df['slope'].astype('category')
df['ca'] = df['ca'].astype('category')
df['thal'] = df['thal'].astype('category')
df['target'] = df['target'].astype('category')

```

```python


df['age'] = df['age'].astype('int8')
df['trestbps'] = df['trestbps'].astype('int16')
df['chol'] = df['chol'].astype('int16')
df['thalach'] = df['thalach'].astype('int16')

```

# Анализ признаков для модели

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Распределение возраста пациентов')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.savefig('/home/igor/LR/eda/histogram_age.png')
plt.show()  
plt.close()
```

## Распределение возраста пациентов показывает, что большинство из них находятся в возрасте от 42 до 70 лет
  

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='sex', hue='target')
plt.title('Частота заболеваний по полу')
plt.xlabel('Пол (0 = Женщина, 1 = Мужчина)')
plt.ylabel('Количество')
plt.legend(title='Заболевание', labels=['Нет', 'Да'])
plt.savefig('/home/igor/LR/eda/countplot_sex_target.png')
plt.show() 
plt.close()
```

## У мужчин чаще встречаются заболевания, но процент заболевших мужчин меньше, нежели процент заболевших женщин

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='target', y='trestbps')
plt.title('Систолическое давление по наличию заболеваний')
plt.xlabel('Заболевание (0 = Нет, 1 = Да)')
plt.ylabel('Систолическое давление (мм рт. ст.)')
plt.savefig('/home/igor/LR/eda/boxplot_trestbps_target.png')
plt.show()  
plt.close()
```

## Пациенты с сердечно-сосудистыми заболеваниями в среднем имеют более низкое систолическое давлениеь

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='target', y='chol')
plt.title('Уровень холестерина по наличию заболеваний')
plt.xlabel('Заболевание (0 = Нет, 1 = Да)')
plt.ylabel('Уровень холестерина (мг/дл)')
plt.savefig('/home/igor/LR/eda/violinplot_chol_target.png')
plt.show() 
plt.close()
```

## Уровень холестерина не сильно информативный признак, но однозначно есть корреляция между высоким уровнем холестирина и наличием заболевания

```python
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Матрица корреляции признаков')
plt.savefig('/home/igor/LR/eda/correlation_heatmap.png')

plt.show()  
```

## Наиболее высокая положительная корреляция наблюдается между 'cp' (тип боли в груди) и 'target' (наличие заболевания),
## 'thalach' (максимальная частота сердечных сокращений) имеет высокую отрицательную корреляцию с 'target',
## что говорит о том, что пациенты с более высокой частотой сердечных сокращений имеют меньший риск заболеваний.
## Уровень холестерина ('chol') имеет положительную корреляцию с целевой переменной, 
## что может указывать на то, что более высокий уровень холестерина ассоциируется с более высоким риском заболеваний.
## 'oldpeak' (депрессия ST) также имеет положительную корреляцию с 'target', что указывает на его важность как  "признака риска заболеваний.
## Другие признаки, такие как 'age' и 'trestbps' (систолическое давление), также имеют умеренные корреляции с 'target', 
## что может указывать на их влияние на риск сердечно-сосудистых заболеваний.


## После анализа можно сделать вывод, что уровень сахара на тощак и кол-во холестирина в крови малоинформотивны и их можно удалить.
